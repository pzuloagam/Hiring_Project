## Hiring Prediction Project

### Overview

This project aims to develop a model to predict hiring decisions based on various candidate attributes. The model will use features such as age, gender, education level, years of experience and others, to determine the likelihood of a candidate being hired. The dataset for this project is sourced from a recent Kaggle entry, which can be found [here](https://www.kaggle.com/datasets/rabieelkharoua/predicting-hiring-decisions-in-recruitment-data/data).

The key objectives of this project are:
1. To preprocess and analyze the dataset to understand the distribution and importance of different features.
2. To develop a model that accurately predicts hiring decisions.
3. To evaluate the performance of the model using appropriate metrics.
4. To provide insights and recommendations based on the model's predictions.


### Data Description and EDA

The dataset for this project includes the following variables:

1. Age: Integer, ranging from 20 to 50 years.
2. Gender: Binary (0: Male, 1: Female).
3. Education Level: Categorical (1: Bachelor's (Type 1), 2: Bachelor's (Type 2), 3: Master's, 4: PhD).
4. Experience Years: Integer, ranging from 0 to 15 years.
5. Previous Companies Worked: Integer, ranging from 1 to 5 companies.
6. Distance From Company: Float, ranging from 1 to 50 kilometers.
7. Interview Score: Integer, ranging from 0 to 100.
8. Skill Score: Integer, ranging from 0 to 100.
9. Personality Score: Integer, ranging from 0 to 100.
10. Recruitment Strategy: Categorical (1: Aggressive, 2: Moderate, 3: Conservative).
10. Hiring Decision: Binary (0: Not hired, 1: Hired).

The dataset contains 1500 entries.
The first 10 columns will be used as features, the last column Hiring Decsision will be the target variable to model.
We will begin with some data exploration and analysis:

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier 
from IPython.core.display import display, HTML
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

warnings.filterwarnings("ignore")
```

```python
# Load the dataset
file_path = r'C:\Users\pzuloaga\Desktop\Machine Learning Project\data\recruitment_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
display(data.head())

# Display the summary statistics of the dataset
display(data.describe())
```

```python

# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Define subsets of categorical and continuous features
categorical_features = ['Gender', 'EducationLevel', 'PreviousCompanies', 'RecruitmentStrategy']
continuous_features = ['Age', 'ExperienceYears', 'DistanceFromCompany', 'InterviewScore', 'SkillScore', 'PersonalityScore']

# Function to plot bar plots for categorical features with specific labels
def plot_categorical(data, features):
    labels = {
        'Gender': ['Male', 'Female'],
        'EducationLevel': ["Bachelor's (Type 1)", "Bachelor's (Type 2)", "Master's", "PhD"],
        'RecruitmentStrategy': ['Aggressive', 'Moderate', 'Conservative']
    }
    
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))
    axs = axs.flatten()
    for i, feature in enumerate(features):
        sns.countplot(x=feature, hue='HiringDecision', data=data, palette='YlGnBu', ax=axs[i])
        axs[i].set_title(f'{feature} vs Hiring Decision')
        axs[i].set_xlabel(feature)
        axs[i].set_ylabel('Count')
        axs[i].legend(title='Hiring Decision', labels=['Not Hired', 'Hired'])
        if feature in labels:
            axs[i].set_xticklabels(labels[feature])
    # Hide any unused subplots
    for j in range(i + 1, len(axs)):
        fig.delaxes(axs[j])
    plt.tight_layout()
    plt.show()

# Function to plot KDE plots for continuous features
def plot_continuous(data, features):
    fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
    axs = axs.flatten()
    for i, feature in enumerate(features):
        sns.kdeplot(data[data['HiringDecision'] == 0][feature], shade=True, label='Not Hired', color='lightblue', ax=axs[i])
        sns.kdeplot(data[data['HiringDecision'] == 1][feature], shade=True, label='Hired', color='orange', ax=axs[i])
        axs[i].set_title(f'{feature} vs Hiring Decision')
        axs[i].set_xlabel(feature)
        axs[i].set_ylabel('Density')
        axs[i].legend(title='Hiring Decision')
    # Hide any unused subplots
    for j in range(i + 1, len(axs)):
        fig.delaxes(axs[j])
    plt.tight_layout()
    plt.show()
```

```python
# Plot for categorical features
plot_categorical(data, categorical_features)
```

```python
# Plot for continuous features
plot_continuous(data, continuous_features)
```

### Model building and training

We weill build and model using five methods studied in class to select the model with the higher accuracy. The methods to analyze are:
1. Random Forest
2. k-Nearest Neighbors
3. Ada Boost
4. Logistic Regression
5. SVC

We will first split the data. Later we will make a function to run all the models.

```python
y = data["HiringDecision"]
x = data.drop("HiringDecision", axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
def training(model):
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    score = accuracy_score(pred, y_test)
    report = classification_report(pred, y_test)
    cm = confusion_matrix(pred, y_test)
    
    return score, report, cm

rfc = RandomForestClassifier()
abc = AdaBoostClassifier()
knn = KNeighborsClassifier()
svc = SVC()
lgr = LogisticRegression()

models = [rfc, abc, knn, gbc, svc, lgr]

names = ["Random Forest", "Ada Boost", " k-Nearest Neighbors ", "Support Vector Machine", "Logistic Regression"]

scores, reports, cms = [], dict(), dict()

for i, j in zip(models, names):
    score, report, cm = training(i)
    scores += [score]
    reports[j] = report
    cms[j] = cm
```

### Results
After training the models we obtained the following results for the accuracy and confussion matrix:

```python
dd = pd.DataFrame({"scores": scores}, index=names)
dd = dd.sort_values("scores", ascending=False)
dd["scores"] = round(dd["scores"], 2)

custom_palette = sns.color_palette('YlGnBu')

# Creating the bar plot
fig, axes = plt.subplots()
sns.barplot(x=dd.index, y=dd["scores"], ax=axes, palette=custom_palette)

for container in axes.containers:
    axes.bar_label(container)
axes.set_xticklabels(axes.get_xticklabels(), rotation=90)
axes.set_yticklabels(())
axes.set_ylabel("")
axes.set_xlabel("")
plt.show()
```

```python
index = 0
fig, axes = plt.subplots(ncols=5, figsize=(20, 4))

for j in range(5):
    sns.heatmap(cms[dd.index[index]], annot=True, fmt='d', ax=axes[j], cmap='YlGnBu')
    axes[j].set_title("{}: {}%".format(dd.index[index], dd.iloc[index, 0]))
    index += 1

plt.tight_layout()
plt.show()
```

We will now use one of the better-performing models to understand feature importance in the final prediction model. Specifically, we will use the Random Forest model, as it provides a feature importance attribute, unlike the SVM model, especially when a non-linear kernel is used.

```python
# Plotting feature importances for the Random Forest model
feature_importances_rf = pd.Series(rfc.feature_importances_, index=features).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances_rf, y=feature_importances_rf.index, palette='YlGnBu')
plt.title('Feature Importances for Random Forest')
plt.xlabel('Importance Value')
plt.ylabel('Feature')
plt.show()
```

### Conclusion

We evaluated five methods to identify the best-performing models. For this dataset, the highest accuracy was achieved with Support Vector Machine (SVM) and Random Forest models, with accuracies ranging from 0.92 to 0.93 in predicting hiring decisions. From the Random Forest results, the most important features were recruitment strategy and interview scores (skills, personality, and overall interview performance). Other significant features included years of experience and education level. Gender was the feature with the least importance in the model.
